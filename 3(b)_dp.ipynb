```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
```

```python
def s1(x1,x2):
    return 1/(1 + np.exp(-(x1 + 50*x2 + 100)))

def s2(x1,x2):
    return 1/(1 + np.exp(-(x1 + 50*x2 - 100)))

def s3(x1,x2):
    return 1/(1 + np.exp(-(50*x1 + x2 + 100)))

def s4(x1,x2):
    return 1/(1 + np.exp(-(50*x1 + x2 - 100)))

def s5(x1,x2):
    return s1(x1,x2) - s2(x1,x2)

def s6(x1,x2):
    return s3(x1,x2) - s4(x1,x2)

def s7(x1,x2):
    return s5(x1,x2) + s6(x1,x2)

def a(x1,x2):
    return 1/(1 + np.exp(-(100*s7(x1,x2) - 200)))
```

```python
x1 = np.arange(-5,5,0.01)
x2 = np.arange(-5,5,0.01)
X, Y = np.meshgrid(x1, x2)
```

```python
Z = s1(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('s1');
```

```python
Z = s2(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x_1')
ax.set_ylabel('x_2')
ax.set_zlabel('s2');
```

```python
Z = s3(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('s3');
```

```python
Z = s4(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('s4');
```

```python
Z = s5(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('s5');
```

```python
Z = s6(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('s6');
```

```python
Z = s7(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('s7');
```

```python
Z = a(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('a');
```
